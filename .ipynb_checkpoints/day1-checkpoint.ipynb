```python
import numpy as np, math

def read_file_lines(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        
        return [line.strip() for line in lines]
```

```python
def get_password(actions, r2=False, debug=0):
    pos = 50
    click = 0
    for a in actions:
        if debug == 2:
            print(pos, '\t', a)
        m = int(a.replace('L', '-').replace('R', ''))
        prev_pos = pos
        
        if r2:
            p = pos
            for i in range(0, abs(m) + 1):
                if debug == 1:
                    print(p, '\t', click)
                if m < 0:
                    p -= 1
                else:
                    p += 1
                if p < 0:
                    p += 100
                elif p >= 100:
                    p -= 100
                if p % 100 == 0:
                    click += 1
            if debug == 1:
                print()
        
        pos = (pos + m) % 100
        if pos == 0 and not r2:
            click += 1
        
    return click

# get_password('R1000'.split(' '), r2=True, debug=1)
# get_password('L68 L30 R48 L5 R60 L55 L1 L99 R14 L82'.split(' '), r2=True, debug=1)
get_password(read_file_lines('day1_input.txt'), r2=True)
```
