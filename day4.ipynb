```python
import numpy as np, math

def read_file_lines(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        
        return [line.strip() for line in lines]
```

```python
stack = [
    '..@@.@@@@.',
    '@@@.@.@.@@',
    '@@@@@.@.@@',
    '@.@@@@..@.',
    '@@.@@@@.@@',
    '.@@@@@@@.@',
    '.@.@.@.@@@',
    '@.@@@.@@@@',
    '.@@@@@@@@.',
    '@.@.@@@.@.'
]

def process_paper(stack, debug=False):
    grid = [list(row) for row in stack]
    grid_out = [row[:] for row in grid]

    c = 0
    for i in range(0, len(grid)):
        for j in range(0, len(grid[0])):
            s = ''
            # Above
            if i > 0:
                s += grid[i - 1][j]
                # Top left
                if j > 0:
                    s += grid[i - 1][j - 1]
                # Top right
                if j < len(grid[0]) - 1:
                    s += grid[i - 1][j + 1]
            # Below
            if i < len(grid) - 1:
                s += grid[i + 1][j]
                # Bottom left
                if j > 0:
                    s += grid[i + 1][j - 1]
                # Bottom right
                if j < len(grid[0]) - 1:
                    s += grid[i + 1][j + 1]
            # Left
            if j > 0:
                s += grid[i][j - 1]
            # Right
            if j < len(grid[0]) - 1:
                s += grid[i][j + 1]

            if grid[i][j] == '@' and s.count('@') < 4:
                grid_out[i][j] = 'x'
                c += 1

    if debug:
        for row in grid_out:
            print("".join(row))
    
    return c, grid_out

# process_paper(stack)
# process_paper(read_file_lines('day4_input.txt'))

def recur_process_paper(stack):
    count, g = process_paper(stack)
    prev_g = None
    while prev_g != g:
        prev_g = g
        c, g = process_paper(g)
        count += c
        
    return count

# recur_process_paper(stack)
recur_process_paper(read_file_lines('day4_input.txt'))
```
