```python
import numpy as np, math

def read_file_lines(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        
        return [line.strip() for line in lines]
```

```python
text = ['123 328  51 64 ',
' 45 64  387 23 ',
'  6 98  215 314',
'*   +   *   +  ']

def process_math(text):
    l = len(text[0].split())
    a = [[] for _ in range(l)]
    
    for row in text:
        r = row.split()
        for i in range(0, l):
            a[i].append(r[i])

    c = 0
    for m in a:
        if m[-1] == '*':
            p = 1
            for e in m[:-1]:
                p *= int(e)
            c += p
        else:
            for e in m[:-1]:
                c += int(e)

    return c

process_math(text)
process_math(read_file_lines('day6_input.txt'))
```

```python
def process_math2(text):
    a = ['' for _ in range(len(text[0]))]
         
    for row in text:    
        for i, ch in enumerate(row):
            a[i] += ch
    
    groups = []
    current = []
    
    for item in a:
        if item.strip() == "":
            if current:
                groups.append(current)
                current = []
        else:
            current.append(item)
    if current:
        groups.append(current)
    
    c = 0

    for m in groups:
        op = m[0][-1]
        m[0] = m[0][:-1]

        if op == '*':
            p = 1
            for e in m:
                p *= int(e)
            c += p
        else:
            for e in m:
                c += int(e)
        
    return c

process_math2(text)
process_math2(read_file_lines('day6_input.txt'))
```
