```python
import numpy as np, math

def read_file_lines(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        
        return [line.strip() for line in lines]
```

```python
bat_bank = ['987654321111111', '811111111111119', '234234234234278', '818181911112111']

def find_max_joltage(bat, n=2):
    b = list(map(int, bat))
    s = 0
    
    for i in range(n):
        idx_max = b.index(max(b[:len(b) - n + i + 1]))
        s += b[idx_max] * 10 ** (n - i - 1)
        b = b[idx_max + 1:]
    
    return s
        
def find_total_joltage(bat_bank, n=2):
    return sum(find_max_joltage(bat, n=n) for bat in bat_bank)
    
find_total_joltage(bat_bank, n=12)
find_total_joltage(read_file_lines('day3_input.txt'), n=12)
```
