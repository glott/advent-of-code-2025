```python
import numpy as np, math, re
from itertools import product

def read_file_lines(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        
        return [line.strip() for line in lines]
```

```python
test_rack = [
    "aaa: you hhh",
    "you: bbb ccc",
    "bbb: ddd eee",
    "ccc: ddd eee fff",
    "ddd: ggg",
    "eee: out",
    "fff: out",
    "ggg: out",
    "hhh: ccc fff iii",
    "iii: out"
]

test_rack2 = [
    "svr: aaa bbb",
    "aaa: fft",
    "fft: ccc",
    "bbb: tty",
    "tty: ccc",
    "ccc: ddd eee",
    "ddd: hub",
    "hub: fff",
    "eee: dac",
    "dac: fff",
    "fff: ggg hhh",
    "ggg: out",
    "hhh: out"
]
```

```python
def expand_until_out(key, d):
    values = d[key]
    result = []

    for v in values:
        if v == 'out':
            result.append("out")
        else:
            result.append(expand_until_out(v, d))

    return result

def process_servers(devices):
    d = {}
    
    for device in devices:
        k, v = device.split(':')
        k = k.strip()
        v = v.strip().split()
        d[k] = v

    expanded = expand_until_out('you', d)

    return str(expanded).count('out')

process_servers(test_rack)
process_servers(read_file_lines('day11_input.txt'))
```

```python
def process_servers2(devices):
    d = {}
    
    for device in devices:
        k, v = device.split(':')
        k = k.strip()
        v = v.strip().split()
        d[k] = v
    
    memo = {}
    def dfs(node, has_fft=False, has_dac=False):
        state = (node, has_fft, has_dac)
        if state in memo:
            return memo[state]

        if node == 'out':
            memo[state] = 1 if has_fft and has_dac else 0
            return memo[state]

        new_has_fft = has_fft or node == 'fft'
        new_has_dac = has_dac or node == 'dac'

        total = 0
        for child in d.get(node, []):
            total += dfs(child, new_has_fft, new_has_dac)

        memo[state] = total
        return total

    return dfs('svr')

process_servers2(test_rack2)
process_servers2(read_file_lines('day11_input.txt'))
```
