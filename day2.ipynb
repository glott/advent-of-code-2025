```python
import numpy as np, math

def read_file_lines(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        
        return [line.strip() for line in lines]
```

```python
test_ids = '11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528, \
446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124'

def div_vals(n):
    vals = []
    for i in range(1, int(n / 2) + 1):
        if n % i == 0:
            vals.append(i)
    return vals

def is_valid(id, mode=1):
    id = str(id)
    if mode == 1:
        if len(id) % 2 == 1:
            return True
    
        h = int(len(id) / 2)
        return id[:h] != id[h:]
    else:
        divs = div_vals(len(id))
    
        for div in divs:
            k = int(len(id) / div)
            is_repeat = True
            prev_split = ''
            for i in range(0, k):
                split = id[i * div:(i + 1) * div]
                if prev_split != split and i != 0:
                    is_repeat = False
                prev_split = split
    
            if is_repeat:
                return False
        return True

def find_invalid_ids(ids, mode=1):
    sum_of_invalid = 0
    for id_range in ids.split(','):
        id_min, id_max = id_range.split('-')
        for id in range(int(id_min), int(id_max) + 1):
            if not is_valid(id, mode=mode):
                sum_of_invalid += id
    return sum_of_invalid

# find_invalid_ids(test_ids, mode=2)
find_invalid_ids(read_file_lines('day2_input.txt')[0], mode=2)
```
