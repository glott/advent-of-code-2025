```python
def find_pairs(boxes, max_steps=10, debug=0):
    boxes = [list(map(int, s.split(','))) for s in boxes]
    dists_grid = [[0] * len(boxes) for _ in range(len(boxes))]
    dists = {}
    
    for i in range(len(boxes)):
        for j in range(i + 1):
            d = math.dist(boxes[i], boxes[j])
            if d != 0:
                dists[str(i) + ',' + str(j)] = d
                dists_grid[i][j] = d

    circuits = []
    steps = 0
    dists = sorted(dists.items(), key=lambda x: x[1])

    prev_k = []
    for k, v in dists:
        if sum(len(row) for row in circuits) == len(boxes):
            b1, b2 = prev_k
            return boxes[int(b1)][0] * boxes[int(b2)][0]
        if steps >= max_steps:
            break
        
        b1, b2 = k.split(',')

        circuit_b1 = None
        circuit_b2 = None

        for i in range(0, len(circuits)):
            if b1 in circuits[i]:
                circuit_b1 = i
            if b2 in circuits[i]:
                circuit_b2 = i

        if circuit_b1 is None and circuit_b2 is None:
            circuits.append([b1, b2])
        elif circuit_b1 == circuit_b2:
            pass
        elif circuit_b1 is not None and circuit_b2 is None:
            circuits[circuit_b1].append(b2)
        elif circuit_b2 is not None and circuit_b1 is None:
            circuits[circuit_b2].append(b1)
        else:
            cb1 = circuits[circuit_b1][:]
            cb2 = circuits[circuit_b2][:]
            temp_circuits = []

            for i in range(0, len(circuits)):
                if i != circuit_b1 and i != circuit_b2:
                    temp_circuits.append(circuits[i])
            temp_circuits.append(cb1 + cb2)
            circuits = temp_circuits[:]
        
        steps += 1
        prev_k = [b1, b2]
        
    for i in range(0, len(boxes)):
        circuit_exists = False
        for c in circuits:
            if str(i) in c:
                circuit_exists = True
        if not circuit_exists:
            circuits.append([str(i)])

    if debug > 0:
        print(sorted(circuits, key=len, reverse=True))
        # for c in sorted(circuits, key=len, reverse=True):
            # s = ''
            # for e in c:
            #     s += str(boxes[int(e)]).replace(' ', '').replace('[', '').replace(']', '') + '\t'
            # print(s)
    if debug > 1:
        for row in dists_grid:
            print(str(row).replace(',','\t'))

    circuit_lengths = sorted([len(inner) for inner in circuits], reverse=True)

    return math.prod(circuit_lengths[:3])

# find_pairs(test_boxes, max_steps=10, debug=1)
find_pairs(read_file_lines('day8_input.txt'), max_steps=10e10, debug=0)
```
