```python
import numpy as np, math, time

def read_file_lines(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        
        return [line.strip() for line in lines]
```

```python
test_ingredients = ['3-5', '10-14', '16-20', '12-18', '', '1', '5', '8', '11', '17', '32']

def process_ranges(ranges):
    ranges.sort(key=lambda x: x[0])
    
    c = 0
    max_up = 0
    for r in ranges:
        low = r[0]
        up = r[1]

        if low <= max_up and up >= max_up:
            c += up - max_up
            max_up = up
        elif low >= max_up:
            c += up - low + 1
            max_up = up
            
    return c
    

def process_ingredients(ingredients):
    idx = ingredients.index('')
    ranges = ingredients[:idx]
    avail = [int(x) for x in ingredients[idx+1:]]

    c = 0

    ranges = [list(map(int, r.split('-'))) for r in ranges]
    for r in ranges:
        for a in avail[:]:
            if a >= r[0] and a <= r[1]:
                c += 1
                avail.remove(a)

    avail_ingredients = process_ranges(ranges)
    return c, avail_ingredients
    

# process_ingredients(test_ingredients)
process_ingredients(read_file_lines('day5_input.txt'))
```
