```python
import numpy as np, math

def read_file_lines(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        
        return [line.strip() for line in lines]
```

```python
test_manifold = [
  ".......S.......",
  "...............",
  ".......^.......",
  "...............",
  "......^.^......",
  "...............",
  ".....^.^.^.....",
  "...............",
  "....^.^...^....",
  "...............",
  "...^.^...^.^...",
  "...............",
  "..^...^.....^..",
  "...............",
  ".^.^.^.^.^...^.",
  "..............."
]

def process_splits(manifold, debug=False):
    m = [list(line) for line in manifold]

    idx_start = m[0].index('S')
    m[0][idx_start] = '|'
    splits = 0
    paths = 1

    for i in range(1, len(m)):
        p = m[i - 1]
        c = m[i]
        
        for j in range(0, len(m[0])):
            if p[j] == '|':
                if c[j] == '.':
                    c[j] = '|'
                elif c[j] == '^':
                    paths += 2
                    has_split = False
                    if j > 0:
                        if c[j - 1] == '.':
                            c[j - 1] = '|'
                            splits += 1
                            has_split = True
                        elif c[j - 1] == '|':
                            paths *= 2
                    if j < len(m[0]):
                        if c[j + 1] == '.':
                            c[j + 1] = '|'
                            if not has_split:
                                splits += 1
                        elif c[j + 1] == '|':
                            paths -= 1

    if debug: 
        for row in m:
            print(''.join(row))
        
    return splits, paths

process_splits(test_manifold)
# process_splits(read_file_lines('day7_input.txt'))
```

```python
def process_splits2(manifold, debug=False):
    m = [list(line) for line in manifold]

    rows = len(m)
    cols = len(m[0])

    ways = [[0] * cols for _ in range(rows)]
    
    idx_start = m[0].index('S')
    m[0][idx_start] = '|'
    ways[0][idx_start] = 1

    splits = 0

    for i in range(0, rows - 1):
        p = m[i]
        
        for j in range(cols):
            if ways[i][j] == 0:
                continue

            count = ways[i][j]
            next_row = i + 1
            char_next = manifold[next_row][j]

            if char_next == '.':
                ways[next_row][j] += count
                m[next_row][j] = '|'
            elif char_next == '^':
                converted = False

                if j > 0:
                    ways[next_row][j - 1] += count
                    if m[next_row][j - 1] == '.':
                        m[next_row][j - 1] = '|'
                        converted = True
                if j < cols - 1:
                    ways[next_row][j + 1] += count
                    if m[next_row][j + 1] == '.':
                        m[next_row][j + 1] = '|'
                        converted = True
                
                if converted:
                    splits += 1
            else:
                pass

    if debug:
        for row in m:
            print(''.join(row))

    total_paths = sum(ways[-1])
    return splits, total_paths
    
process_splits2(test_manifold, True)
process_splits2(read_file_lines('day7_input.txt'))
```
