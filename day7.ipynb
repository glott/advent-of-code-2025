```python
import numpy as np, math

def read_file_lines(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
        
        return [line.strip() for line in lines]
```

```python
test_manifold = [
  ".......S.......",
  "...............",
  ".......^.......",
  "...............",
  "......^.^......",
  "...............",
  ".....^.^.^.....",
  "...............",
  "....^.^...^....",
  "...............",
  "...^.^...^.^...",
  "...............",
  "..^...^.....^..",
  "...............",
  ".^.^.^.^.^...^.",
  "..............."
]

def process_splits(manifold, debug=False):
    m = [list(line) for line in manifold]

    idx_start = m[0].index('S')
    m[0][idx_start] = '|'
    splits = 0

    for i in range(1, len(m)):
        p = m[i - 1]
        c = m[i]
        
        for j in range(0, len(m[0])):
            if p[j] == '|':
                if c[j] == '.':
                    c[j] = '|'
                elif c[j] == '^':
                    has_split = False
                    if j > 0:
                        if c[j - 1] == '.':
                            c[j - 1] = '|'
                            splits += 1
                            has_split = True
                    if j < len(m[0]):
                        if c[j + 1] == '.':
                            c[j + 1] = '|'
                            if not has_split:
                                splits += 1

    if debug: 
        for row in m:
            print(''.join(row))
        
    return splits

process_splits(test_manifold)
# process_splits(read_file_lines('day7_input.txt'))
```

```python
def process_splits2(manifold, debug=False):
    m = [list(line) for line in manifold]
    
    idx_start = m[0].index('S')
    m[0][idx_start] = '|'

    splits = 0
    
    paths = [[0] * len(m[0]) for _ in range(len(m))]
    paths[0][idx_start] = 1

    for i in range(0, len(m) - 1):
        n = m[i + 1]
        np = paths[i + 1]
        
        for j in range(len(m[0])):
            if paths[i][j] == 0:
                continue

            count = paths[i][j]
            char_next = manifold[i + 1][j]

            if char_next == '.':
                np[j] += count
                n[j] = '|'
            elif char_next == '^':
                has_split = False

                if j > 0:
                    np[j - 1] += count
                    if n[j - 1] == '.':
                        n[j - 1] = '|'
                        has_split = True
                if j < len(m[0]) - 1:
                    np[j + 1] += count
                    if n[j + 1] == '.':
                        n[j + 1] = '|'
                        has_split = True
                
                if has_split:
                    splits += 1
            else:
                pass

    if debug:
        for row in m:
            print(''.join(row))

    total_paths = sum(paths[-1])
    return splits, total_paths
    
process_splits2(test_manifold)
process_splits2(read_file_lines('day7_input.txt'))
```
